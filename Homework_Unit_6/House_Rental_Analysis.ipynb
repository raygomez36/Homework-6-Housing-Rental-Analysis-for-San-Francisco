# Homework 5 PyViz

```python
import pandas as pd
import numpy as np
import hvplot.pandas

%matplotlib inline
```

```python
#Creating Dataframe
sfnc_data = pd.read_csv('sfo_neighborhoods_census_data.csv', usecols=['year', 'sale_price_sqr_foot', 
                                                                      'housing_units', 'gross_rent'])
sfnc_data.head()
```

```python
#Calculating mean grouping by year
sfnc_bar = sfnc_data.groupby("year").mean()
sfnc_bar
```

```python
#Plotting Housing Unit by Year
sfnc_bar.hvplot.bar(
title='Housing Units in SanFrancisco from 2010 to 2016',
x = 'year',
y = 'housing_units',
xlabel = 'Year',
ylabel = 'Housing Units'
).opts(yformatter='%.0f')
```

```python
# What’s the overall trend in housing units over the period that you’re analyzing?
# According to the chart very little has changed during the years listed on the chart.
```

# Compare the Average Sale Prices by Neighborhood

```python
#Creating DataFrame
df_data = pd.read_csv('sfo_neighborhoods_census_data.csv')
df_data.head()
```

```python
#Calculating mean and grouping by year & neighborhood
neighborhood_pricing_per_year = df_data.groupby(['year', 'neighborhood']).mean()
neighborhood_pricing_per_year
```

```python
#Filitering Dataframe
neighborhood_pricing = neighborhood_pricing_per_year[['sale_price_sqr_foot', 'gross_rent']]
neighborhood_pricing.head()
```

```python
#Plotting Neighborhood Pricing 
neighborhood_pricing.hvplot.line(

    x='year',
    xlabel ='Year',
    ylabel = 'Sale Price Yearly & Gross Rent',
    groupby='neighborhood',
    title = 'Sale Price Per Square Foot and Average Gross Rent - 2010-2016- By Neighborhood'
)
```

# Build an Interactive Neighborhood Map

```python
#Created DataFrame with sfo_neighborhoods_census_data.csv
sfo_data_df = pd.read_csv('sfo_neighborhoods_census_data.csv')
sfo_data_df
```

```python
#Creating DataFrame from neighborhoods_coordinates.csv
neighborhood_locations_df = pd.read_csv('neighborhoods_coordinates.csv', index_col='Neighborhood')
neighborhood_locations_df.head()
```

```python
#Aggregating DataFrame with mean
all_neighborhood_info_df = sfo_data_df.groupby('neighborhood').mean()
all_neighborhood_info_df
```

```python
#Concat both Dataframes
all_neighborhoods_df = pd.concat([neighborhood_locations_df, all_neighborhood_info_df], axis='columns',
                                sort=False)
all_neighborhoods_df = all_neighborhoods_df.dropna()
all_neighborhoods_df
```

```python
#Plotting HV Plot for Interactive Map
all_neighborhoods_df.hvplot.points(
    'Lon',
    'Lat',
    geo=True,
    color='gross_rent',
    size='sale_price_sqr_foot',
    tiles='OSM',
    hover_cols='all',
    frame_width=700,
    frame_height=500,
    title='Intecactive Map Plotting San Francisco'
)
```
